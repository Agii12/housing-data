```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r'housing_data.xlsx')
df.head(10)
```

```python
df.describe()
```

```python
print("Missing values count:\n", df.isnull().sum())
```

```python
#first visual
#average home price
plt.figure(figsize=(10,6))
plt.plot(df.index, df['Average Home Price'], color='green', label='Average Home Price')
plt.title('Average Home Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price in USD')
plt.grid(True)
plt.legend()
plt.show()         
```

```python
#second visual 
#interest rate
#a line plot of interest rate during that same period
plt.figure(figsize=(10,6))
plt.plot(df.index,df['Interest Rate (%)'], color='red', label='Interest Rate (%)')
plt.title('Interest Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Interest Over Time')
plt.grid()
plt.legend()
plt.show()
```

```python
#third visual 
#sales volume
plt.figure(figsize=(10,6))
plt.plot(df.index,df['Sales Volume'], color='blue', label='Sales Volume')
plt.title('Sales Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Volume')
plt.grid()
plt.legend()
plt.show()
```

```python
#4th visual
#oHousing inventory levels
plt.figure(figsize=(10,6))
plt.plot(df.index,df['Inventory'], color='green', label='Inventory')
plt.title('Inventory Over Time')
plt.xlabel('Date')
plt.ylabel('Inventory')
plt.grid()
plt.legend()
plt.show()
```

```python
#2.	Examine correlations between these factors to understand how they might influence each other
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Heatmap on hpousing Data ")
plt.show()
```

```python
df.corr(method='pearson')
```

```python
#plot sales volumn and price trend
plt.figure(figsize = (14,6))

#plot sales volumn
plt.subplot(2, 1, 1)
plt.plot(df.index,df['Sales Volume'], color='blue', label='Sales Volume')
plt.title('Monthly Sales Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Volume')
plt.grid()
plt.legend()

# plot price trend
plt.subplot(2, 1, 2)
plt.plot(df.index, df['Average Home Price'], color='green', label='Average Home Price')
plt.title('Monthly Average Home Price Over Time')
plt.xlabel('Date')
plt.ylabel('Average Home Price')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
